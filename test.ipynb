```python
# Set up user authentication
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=client_id,
    client_secret=client_secret,
    redirect_uri="https://127.0.0.1",
    scope="playlist-modify-public playlist-modify-private user-read-private"
))

artist_name = "Tommy Guerrero"
user = sp.current_user()
user_id = user['id']
print(f"Logged in as: {user['display_name']}")

# Search for the artist
print(f"\nSearching for artist: {artist_name}")
results = sp.search(q=f'artist:{artist_name}', type='artist', limit=1)

if not results['artists']['items']:
    print(f"Artist '{artist_name}' not found!")

artist = results['artists']['items'][0]
artist_id = artist['id']
print(f"Found artist: {artist['name']} (Followers: {artist['followers']['total']:,})")

# Get artist's top tracks
print(f"\nGetting top tracks for {artist['name']}...")
top_tracks = sp.artist_top_tracks(artist_id, country='US')


tracks_to_add = top_tracks['tracks'][:2]  # Top 2 songs

# Display the tracks we're going to add
print(f"\nTop {len(tracks_to_add)} tracks:")
track_uris = []
for i, track in enumerate(tracks_to_add, 1):
    print(f"{i}. {track['name']} - {track['album']['name']}")
    track_uris.append(track['uri'])

# Create or find playlist
playlist_name = 'My API Playlist'

# Check if playlist already exists
playlists = sp.user_playlists(user_id)
existing_playlist = None
for playlist in playlists['items']:
    if playlist['name'] == playlist_name:
        existing_playlist = playlist
        break

if existing_playlist:
    print(f"\nUsing existing playlist: {playlist_name}")
    playlist_id = existing_playlist['id']
else:
    print(f"\nCreating new playlist: {playlist_name}")
    new_playlist = sp.user_playlist_create(
        user=user_id,
        name=playlist_name,
        public=True,
        description=f"Top songs by {artist['name']} - Created via Spotify API"
    )
    playlist_id = new_playlist['id']

# Add tracks to playlist
print(f"\nAdding {len(track_uris)} tracks to playlist...")
sp.playlist_add_items(playlist_id, track_uris)

# Get playlist URL
playlist_info = sp.playlist(playlist_id)
playlist_url = playlist_info['external_urls']['spotify']

print(f"\n✅ Success!")
print(f"Added {len(track_uris)} tracks to playlist '{playlist_name}'")
print(f"Playlist URL: {playlist_url}")


```
